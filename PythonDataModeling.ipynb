```python
# Plotting the normal distribution of Anomaly Scores per Attack Type

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

plt.figure(figsize=(12, 6))

for attack_type in df['Attack Type'].unique():
    subset = df[df['Attack Type'] == attack_type]
    mean = subset['Anomaly Scores'].mean()
    std = subset['Anomaly Scores'].std()
    x = np.linspace(mean - 3*std, mean + 3*std, 100)
    plt.plot(x, norm.pdf(x, mean, std), label=attack_type)

plt.title('Normal Distribution of Anomaly Scores per Attack Type')
plt.xlabel('Anomaly Scores')
plt.ylabel('Density')
plt.legend(title='Attack Type')
plt.show()
```

```python
# Plot the density of Anomaly Scores per Attack Type
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
ax = sns.kdeplot(data=df, x='Anomaly Scores', hue='Attack Type')

# Manually create the legend
handles, labels = ax.get_legend_handles_labels()
if handles and labels:
    ax.legend(handles=handles, labels=labels, title='Attack Type')

ax.set_title('Normal Distribution of Anomaly Scores per Attack Type')
ax.set_xlabel('Anomaly Scores')
ax.set_ylabel('Density')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter for high-severity attacks without proxy
filtered_df = df[(df['Severity Level'] == 'High') & (df['Proxy Information'].isnull())]

# Get protocol distribution
protocol_distribution = filtered_df['Protocol'].value_counts()

# Plot using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=protocol_distribution.index, y=protocol_distribution.values, palette="viridis")

# Add title and labels
plt.title('Protocol Distribution for High-Severity Attacks Without Proxy', fontsize=16)
plt.xlabel('Protocol', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.ylim(2000, protocol_distribution.max() + 10)  # Starting from 2000, and extending to the max + 1000 for the upper limit

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert 'Timestamp' to a datetime format for time-based analysis
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Filter data for the holiday season (from November 25 to December 31)
holiday_data = df[(df['Timestamp'].dt.month == 11) & (df['Timestamp'].dt.day >= 25) |
                  (df['Timestamp'].dt.month == 12)]

# Group the data by week number (using Timestamp.dt.week) and count alerts/warnings for each week
holiday_data['Week'] = holiday_data['Timestamp'].dt.isocalendar().week
alerts_weekly = holiday_data.groupby('Week')['Alerts/Warnings'].count()

# Plotting the trend of alerts/warnings per week during the holiday season
plt.figure(figsize=(12, 6))
alerts_weekly.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Week Number', fontsize=12)
plt.ylabel('Number of Alerts/Warnings', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

proxy_data = df[df['Proxy Information'].notnull()]

proxy_data['Is Malicious'] = proxy_data[['Malware Indicators', 'IDS/IPS Alerts']].notnull().any(axis=1)

malicious_count = proxy_data['Is Malicious'].sum()
non_malicious_count = len(proxy_data) - malicious_count

# Data for the pie chart
labels = ['Malicious Proxy Traffic', 'Non-Malicious Proxy Traffic']
sizes = [malicious_count, non_malicious_count]
colors = ['red', 'green']

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Proportion of Malicious Proxy-Based Traffic', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Filter rows where Anomaly Scores are available
df_filtered = df[df['Anomaly Scores'].notnull()]

# Ensure the severity levels are ordered
severity_order = ['Low', 'Medium', 'High']
df_filtered['Severity Level'] = pd.Categorical(df_filtered['Severity Level'], categories=severity_order, ordered=True)

# Group by Severity Level and calculate the average Anomaly Score
heatmap_data = df_filtered.groupby('Severity Level')['Anomaly Scores'].mean().reset_index()

# Pivot the data for heatmap compatibility
heatmap_data_pivot = pd.pivot_table(heatmap_data, values='Anomaly Scores', index=['Severity Level'])

# Plotting the heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(heatmap_data_pivot, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)

# Adding labels and title
plt.title('Average Anomaly Scores by Severity Level', fontsize=14)
plt.ylabel('Severity Level')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Filter the dataset for rows where 'Malware Indicators' has meaningful values
df_filtered = df[df['Malware Indicators'].notnull()]

# Create a new column that indicates whether IoC was detected or not
df_filtered['IoC Detected'] = df_filtered['Malware Indicators'].apply(lambda x: 'Detected' if x == 'IoC Detected' else 'Not Detected')

# Group by 'IoC Detected' and calculate the count of alerts
heatmap_data = df_filtered.groupby('IoC Detected').size().reset_index(name='Alert Count')

# Pivot the data for heatmap compatibility
heatmap_data_pivot = pd.pivot_table(heatmap_data, values='Alert Count', index=['IoC Detected'])

# Plotting the heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(heatmap_data_pivot, annot=True, fmt="d", cmap='coolwarm', cbar=True)

# Adding labels and title
plt.title('Number of Alerts Triggered by IoC Detection Status', fontsize=14)
plt.ylabel('IoC Detection Status')
plt.xlabel('Alert Count')
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# The data set is loaded into the dataframe
df = pd.read_csv('cybersecurity_attacks.csv')

# Count the number of each attack
counts = df['Attack Type'].value_counts()

# Compute the percentage for the Attack Type
percentage = (counts / counts.sum()) * 100

# Visualize the result in a pie chart
plt.figure(figsize=(8, 8))
plt.pie(percentage, labels=percentage.index, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'], startangle=90, autopct='%1.1f%%')
plt.title('Spread of Attack Type')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming df contains the dataset

# Filter rows where Anomaly Scores are not null
df_filtered = df[df['Anomaly Scores'].notnull()]

# Create a new column to classify rows into 'Alert Data' and 'Non-Alert Data'
df_filtered['Alert Type'] = df_filtered['IDS/IPS Alerts'].apply(lambda x: 'Alert Data' if x == 'Alert Data' else 'Non-Alert Data')

# Group by 'Alert Type' and calculate the mean anomaly score
heatmap_data = df_filtered.groupby('Alert Type')['Anomaly Scores'].mean().reset_index()

# Pivot the data for heatmap compatibility
heatmap_data_pivot = pd.pivot_table(heatmap_data, values='Anomaly Scores', index=['Alert Type'])

# Plotting the heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(heatmap_data_pivot, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)

# Adding labels and title
plt.title('Mean Anomaly Score for IDS/IPS Alerts and Non-Alerts', fontsize=14)
plt.ylabel('Alert Type')
plt.tight_layout()
plt.show()
```
